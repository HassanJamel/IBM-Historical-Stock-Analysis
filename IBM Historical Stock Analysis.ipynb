# 📊 IBM Historical Stock Analysis (2000-2026)
## Comprehensive Exploratory Data Analysis & Financial Storytelling

### **Project Overview**
This notebook performs a deep dive into the historical performance of International Business Machines Corp. (IBM) over a 26-year span. We explore price trends, volatility patterns, and the underlying business transformations that have shaped IBM's journey from a hardware giant to a cloud and AI leader.

### **Key Objectives**
- **EDA & Preprocessing:** Clean and prepare historical OHLCV data.
- **Visual Insights:** Use Seaborn and Plotly for static and animated visualizations.
- **Financial Metrics:** Analyze moving averages, daily returns, and risk profiles.
- **Storytelling:** Map stock performance to real-world corporate milestones.
- **Problem Solving:** Identify core bottlenecks and propose strategic solutions.


```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import warnings
from datetime import datetime

# Settings
sns.set_theme(style="whitegrid", palette="muted")
plt.rcParams['figure.figsize'] = (12, 6)
warnings.filterwarnings('ignore')

print("✅ Environment Ready")
```

```
# Load the dataset
file_path = 'IBM.csv'
df = pd.read_csv(file_path)

# Convert Date to datetime objects
df['Date'] = pd.to_datetime(df['Date'])
df = df.sort_values('Date')

# Basic Overview
print(f"Dataset Shape: {df.shape}")
df.head()
```

### **4. Data Preprocessing & Feature Engineering**
Transforming raw data into actionable financial features:
1. **Daily Returns:** Percentage change in closing price.
2. **Volatility:** Rolling standard deviation of returns.
3. **Moving Averages (MA):** 50-day and 200-day trends.


```
# Feature Engineering
df['Daily_Return'] = df['Close'].pct_change() * 100
df['MA50'] = df['Close'].rolling(window=50).mean()
df['MA200'] = df['Close'].rolling(window=200).mean()
df['Volatility'] = df['Daily_Return'].rolling(window=30).std()

# Drop initial NaNs from rolling windows
df.dropna(inplace=True)

print("✅ Preprocessing Complete")
df.describe()
```

### **5. Understanding Data (Composition, Distribution, Comparison & Relationships)**
We analyze the data across four dimensions to understand its fundamental properties.


```
def plot_univariate_analysis(df):
    """
    Analyzes the distribution of Close Prices and Daily Returns.
    - Histogram: Shows frequency distribution.
    - Boxplot: Identifies outliers and quartiles.
    """
    fig, axes = plt.subplots(1, 2, figsize=(15, 5))
    
    sns.histplot(df['Close'], kde=True, ax=axes[0], color='blue')
    axes[0].set_title('Distribution of Closing Prices')
    
    sns.boxplot(x=df['Daily_Return'], ax=axes[1], color='green')
    axes[1].set_title('Daily Returns Outliers')
    
    plt.tight_layout()
    plt.show()

plot_univariate_analysis(df)
```

### **6. Patterns, Trends, and Outliers**
Analyzing the 26-year growth trajectory and identifying key market anomalies.


```
# High-Quality Trend Analysis
fig = go.Figure()

fig.add_trace(go.Scatter(x=df['Date'], y=df['Close'], name='Close Price', line=dict(color='royalblue', width=1.5)))
fig.add_trace(go.Scatter(x=df['Date'], y=df['MA200'], name='200-Day MA (Long-term)', line=dict(color='firebrick', width=2, dash='dash')))

fig.update_layout(title='IBM 26-Year Price Trend with 200-Day Moving Average',
                  xaxis_title='Year', yaxis_title='Price (USD)',
                  template='plotly_dark', hovermode='x unified')
fig.show()
```

### **7. Animated Financial Journey**
Visualizing the evolution of IBM's price action over the decades.


```
# Yearly Performance Animation
df['Year'] = df['Date'].dt.year
yearly_avg = df.groupby('Year')[['Close', 'Volume']].mean().reset_index()

fig = px.scatter(yearly_avg, x='Year', y='Close', size='Volume', color='Close',
                 hover_name='Year', log_x=False, size_max=60,
                 title='IBM: Average Yearly Price & Volume Evolution (Animated)',
                 animation_frame='Year', animation_group='Year',
                 range_x=[1999, 2027], range_y=[0, 350],
                 template='plotly_white')

fig.show()
```

### **8. Professional Interactive Dashboard**
A consolidated view of IBM's health metrics.


```
def create_dashboard(df):
    fig = make_subplots(rows=2, cols=2, 
                        subplot_titles=('Price Trend', 'Volatility Clusters', 'Daily Returns Hist', 'Volume spikes'),
                        vertical_spacing=0.12)

    # 1. Price
    fig.add_trace(go.Scatter(x=df['Date'], y=df['Close'], name='Price'), row=1, col=1)
    
    # 2. Volatility
    fig.add_trace(go.Scatter(x=df['Date'], y=df['Volatility'], name='Volatility', line=dict(color='orange')), row=1, col=2)
    
    # 3. Returns
    fig.add_trace(go.Histogram(x=df['Daily_Return'], name='Returns Hist', marker_color='green'), row=2, col=1)
    
    # 4. Volume
    fig.add_trace(go.Bar(x=df['Date'], y=df['Volume'], name='Volume', marker_color='purple'), row=2, col=2)

    fig.update_layout(height=800, width=1100, title_text="IBM Interactive Market Dashboard", template='plotly_dark')
    fig.show()

create_dashboard(df)
```

### **9. Data Storytelling & Root Problem Analysis**

#### **The Narrative**
IBM entered the 2000s as a hardware titan. The dataset shows the massive impact of the **Dot-Com Crash (2000-2002)**, followed by a decade of steady growth as it pivoted toward services. However, the 2010s show a "Flattening Period" where IBM struggled with legacy systems while younger competitors dominated Cloud and Mobile. The 2020-2026 data reflects a "Renaissance" driven by Hybrid Cloud and AI.

#### **Core Root Problem**
**"Transition Inertia"**: IBM's core struggle was the slow abandonment of high-margin legacy hardware for the high-growth, but initially lower-margin, cloud services ecosystem.

#### **Problem Mapping**
- **Cause:** Over-reliance on traditional mainframe and consulting revenue.
- **Failure:** Delayed entry into the public cloud market compared to AWS and Azure.
- **Outcome:** Sub-par stock performance between 2013-2019 compared to the S&P 500 Tech index.
